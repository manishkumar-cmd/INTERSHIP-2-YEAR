```python
x=np.array([1,3,5])
plt.pie(x)
plt.show()
```

```python
x=np.array([1,3,5])
mylabels=np.array(["apple","orange","banana"])
plt.pie(x,labels=mylabels,startangle=50)
plt.show()
```

```python
x=np.array([1,3,5])
mylabels=np.array(["apple","orange","banana"])
myexplode=np.array([0.3,0,0])
plt.pie(x,labels=mylabels,explode=myexplode,startangle=50)
plt.show()
```

```python
x=np.array([1,3,5,6])
mylabels=np.array(["apple","orange","banana","mango"])
myexplode=np.array([0.2,0,0,0])
plt.pie(x,labels=mylabels,explode=myexplode,startangle=50)
plt.show()
```
