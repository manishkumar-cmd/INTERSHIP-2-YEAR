```python
import pandas as pd
df = pd.read_csv('laptopPrice.csv')
asus_laptops = df[df['brand'] == 'ASUS']
print(asus_laptops.to_string(index=False))
```

laptop with minimum 256 gb ssd and windows 64 bit  and 8 gb ram minimun 10 gen processor laptops with minimum 3 star and under 100000

```python
filtered_laptops = df[
    (df['ssd'].str.replace(' GB', '').astype(int) >= 256) &
    (df['os'] == 'Windows') &
    (df['os_bit'] == '64-bit') &
    (df['ram_gb'].str.replace('GB', '').astype(int) >= 8) &
    (df['processor_gnrtn'].str.contains('10th|11th|Not Available', case=False, na=False)) & 
    (df['processor_name'].str.contains('Ryzen', case=False)) &
    (df['rating'].str.contains('3|4|5', regex=True)) &
    (df['Price'] < 100000)]
print(filtered_laptops.to_string(index=False))
```

msi laptop with minimum 1 year warrenty

```python
import pandas as pd

df = pd.read_csv('laptopPrice.csv')

msi_laptops = df[
    (df['brand'] == 'MSI') & 
    (df['warranty'].str.contains('1 year|2 years|3 years', case=False, na=False))
]

print(msi_laptops.to_string(index=False))
```

```python
import pandas as pd

df = pd.read_csv('laptopPrice.csv')

msi_laptops = df[
    (df['brand'] == 'MSI') & 
    (df['warranty'].str.contains('1 year|2 years|3 years', case=False, na=False)) &
    (df['rating'].str.contains('4 stars|5 stars', case=False))
]

print(msi_laptops.to_string(index=False))
```

```python
import pandas as pd

df = pd.read_csv('laptopPrice.csv')

msi_i7_laptops = df[
    (df['brand'] == 'MSI') &
    (df['processor_name'].str.contains('Core i7', case=False)) &
    (df['warranty'].str.contains('1 year|2 years|3 years', case=False, na=False)) &
    (df['rating'].str.contains('4 stars|5 stars', case=False))
]

print(msi_i7_laptops.to_string(index=False))
```

```python
import pandas as pd
df = pd.read_csv('laptopPrice.csv')
msi_i7_laptops = df[
    (df['brand'] == 'MSI') &
    (df['processor_name'].str.contains('Core i7|Core i9|Ryzen 7|Ryzen 9', case=False)) &
    (df['warranty'].str.contains('1 year|2 years|3 years', case=False, na=False)) &
    (df['rating'].str.contains('4 stars|5 stars', case=False)) &
    (df['Price'] < 100000)
]
print(msi_i7_laptops.to_string(index=False))
msi_i7_laptops.to_csv('msi_high_end_under_1lakh.csv', index=False)
```
